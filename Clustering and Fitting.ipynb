```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\Other\OneDrive - University of Hertfordshire\ADS1\As2\dataset air.zip"  # Replace with the correct path
data = pd.read_csv(file_path)
```

```python
def create_line_plot(data, columns, title):
    plt.figure(figsize=(14, 8))  # Larger figure size for readability
    for col in columns:
        mean = data[col].mean()
        std = data[col].std()
        plt.plot(data.index, data[col], label=col, linewidth=2)  # Thicker lines for clarity
    plt.title(title, fontsize=16)  # Larger title font
    plt.xlabel('Index', fontsize=14)  # Larger x-axis label
    plt.ylabel('Values', fontsize=14)  # Larger y-axis label
    plt.xticks(fontsize=12)  # Readable x-ticks
    plt.yticks(fontsize=12)  # Readable y-ticks
    plt.legend(fontsize=12)  # Clear legend with larger font
    plt.grid(True, linestyle='--', alpha=0.7)  # Dotted grid for clarity
    plt.tight_layout()  # Ensure nothing overlaps
    plt.show()
    print("\nStatistical Summary:")
    print(data[columns].describe())
    print("\nSkewness:\n", data[columns].skew())
    print("\nKurtosis:\n", data[columns].kurtosis())

# Example usage:
create_line_plot(data, columns=['Temperature', 'Humidity', 'PM2.5'], 
                       title="Line Plot with Statistical Summaries")

plt.savefig('line_plot.png')  # Change 'line_plot.png' to your desired file path
plt.close()
```

```python
def create_bar_plot(data, x_column, y_column, title):
    plt.figure(figsize=(14, 8))  # Larger figure size for readability
    sns.barplot(x=x_column, y=y_column, data=data, hue=y_column, palette="viridis", legend=False)
    plt.title(title, fontsize=16)  # Larger font for the title
    plt.xlabel(x_column, fontsize=14)  # Larger x-axis label
    plt.ylabel(y_column, fontsize=14)  # Larger y-axis label
    plt.xticks(fontsize=12, rotation=45, ha='right')  # Rotated x-axis labels
    plt.yticks(fontsize=12)  # Readable y-axis labels
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # Grid lines for better visual comparison
    plt.tight_layout()  # Ensure nothing overlaps
    plt.show()
   # Check if the y_column is numeric
    if data[y_column].dtype in ['float64', 'int64']:
        print(f"\nStatistical Summary for {y_column}:")
        print(data[y_column].describe())
        print(f"\nSkewness: {data[y_column].skew():.2f}")
        print(f"Kurtosis: {data[y_column].kurtosis():.2f}")
    else:
        print(f"\n{y_column} contains non-numeric data. Statistical summaries skipped.")

# Example usage:
create_bar_plot(data, x_column='Population_Density', y_column='Air Quality', 
                      title="Bar Plot Comparing Population Density vs Air Quality")


plt.savefig('bar_plot.png')  # Change 'bar_plot.png' to your desired file path
plt.close()
```

```python
def create_heatmap(data, title):
    plt.figure(figsize=(14, 10))  # Larger figure size for readability
    correlation_matrix = data.corr(numeric_only=True)  # Compute correlation for numeric columns only
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", 
                linewidths=0.5, annot_kws={"fontsize": 10})  # Readable annotations
    plt.title(title, fontsize=16)  # Larger title font
    plt.xticks(fontsize=12, rotation=45, ha='right')  # Readable and rotated x-axis labels
    plt.yticks(fontsize=12, rotation=0)  # Readable y-axis labels
    plt.tight_layout()  # Ensure nothing overlaps
    plt.show()
    print("\nStatistical Summary:")
    print(data.describe())
    print("\nSkewness:\n", data.skew(numeric_only=True))
    print("\nKurtosis:\n", data.kurtosis(numeric_only=True))

# Example usage:
create_heatmap(data, title="Heatmap with Correlation Matrix and Statistics")


plt.savefig('heatmap.png')  # Change 'heatmap.png' to your desired file path
plt.close()
```
